# Sales Analysis using pandas


```python
import os
import pandas as pd
```

```python
path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
    
```

```python
all_data = pd.read_csv("all_data_copy.csv")
all_data.head()
```

#### Q1) Which month was the best for sales? How much  dough was made?

```python
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
#Method 2 for creating the float columns
#all_data["Month"] = pd.to_datetime(all_data["Order Date"]).dt.month
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

```python
outcome = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

months = range(1,13)
plt.bar(months,outcome['Sales'])
plt.xticks(months)
plt.xlabel('months in no.')
plt.ylabel('Sales in ($)')
plt.show()
```

#### Q2) What city has the most sales?

```python
all_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(" ")[1]
    



all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x)+' ('+get_state(x) +')')

all_data.head()
```

```python
outcome = all_data.groupby('City').sum()
```

```python
outcome
```

```python
#Reutrning the same order of cities as present in all_data.groupby('City')
cities = [city for city,df in all_data.groupby('City')]


plt.bar(cities,outcome['Sales'])
plt.xticks(cities, rotation = 'vertical', size = 9)
plt.xlabel('Cities')
plt.ylabel('Sales in ($)')
plt.show()
```

#### Q3) What time is perfect for putting up ads so that the likelihood of customer buying products increases?

```python
all_data.head()
```

```python
# Aggregate the count of orders placed (grouped by time provided in order date) over a 24hr period
```

```python
all_data['Hour']= pd.to_datetime(all_data['Order Date']).dt.hour
all_data['Minute']=pd.to_datetime(all_data['Order Date']).dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
outcome = all_data.groupby('Hour').count()
```

```python
outcome
```

```python
hrs = [ hour for hour,df in all_data.groupby('Hour')]



plt.plot(hrs,outcome['Count'])
plt.xticks(hrs)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()

```

#### What products are most often sold together?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
```

```python
df.head()
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
```

```python
df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(15):
    print(key,value)
```
